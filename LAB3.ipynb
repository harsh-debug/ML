```
import numpy as np
import scipy.stats as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

#**Question 1:**

##**Q1 --- Part:1) Bernoulli Distribution**


```
prob_of_success = np.array([0.2,0.4,0.6,0.9])
x = np.array([0,1])
for p in prob_of_success:
  plt.plot(x, sp.bernoulli.pmf(x,p), 'o',ms=8, label="p = "+str(p))
  plt.vlines(x, 0, sp.bernoulli.pmf(x, p),linewidth=3, alpha=0.2)
  print("For p = {} : \n Mean = {}, Variance = {}\n".format(p,sp.bernoulli.mean(p),sp.bernoulli.var(p)))
plt.legend() #plt.legend(['p = 0.2', 'p = 0.4', 'p = 0.6','p = 0.9'])
plt.xlim(-1,2)
plt.xlabel('Data Points')
plt.ylabel('Probability')
plt.title('Probability Mass Function(PMF) of Bernoulli Distribution')
plt.show()

```

```
plt.figure(figsize=(7,7))
x =np.arange(-1,2,0.2)
p=0.2                           #Probability of Success
y = sp.bernoulli.cdf(x,p)

plt.step(x,y,label="p = 0.2")
p=0.6                           #Probability of Success
y = sp.bernoulli.cdf(x,p)
plt.step(x,y,'-.',label="p = 0.6")

plt.legend() #plt.legend(['p = 0.2','p = 0.6'])
plt.xlim(-1,2)
plt.xlabel('Data Points')
plt.ylabel('Probability')
plt.title('Cumulative Distribution Function (CDF) of Bernoulli Distribution')
plt.show()
```

##**Q1 --- Part:2) Multinoulli Distribution**

```
n=1 # Number of experiment      ## For n=1 , multinomial is multinoulli
pval=[0.2, 0.1, 0.3, 0.4]       # Prob of success for ith random variable
x=np.random.multinomial(n=1,pvals=pval,size=10)
print("Multinoulli :\n Mean = {} and Variance = {} ".format(np.mean(x),np.var(x)))
y = np.arange(1,len(pval)+1)
plt.plot(y,pval,'o')
plt.xlabel('ith Random Variable')
plt.ylabel('Probability of X_i = 1')
plt.title('Probability Mass Function(PMF) of Multinoulli Distribution')
plt.vlines(y, 0, pval,linewidth=3, alpha=0.2)
plt.show()

cdf_prob=[]
sum=0;
for i in pval:
  sum += i
  cdf_prob.append(sum)
plt.step(y,cdf_prob,'-.',where='post')
plt.ylim(0)
plt.title('Cumulative Distribution Function (CDF) of Multinoulli Distribution')
plt.show()
```

##**Q1 --- Part:3) Normal Distribution**

```
x =np.arange(-13,13,0.001)
plt.figure(figsize=(10,10))

#### Keeping Mean constant = 0 and varying standard deviation
mean = 0
standard_deviation = [2,3,4,5]
for sd in standard_deviation:
  plt.plot(x,sp.norm.pdf(x,mean,sd),label="Mean = {} & SD = {}".format(mean,sd))
plt.legend()
plt.title("Probability Density Function(PDF) of Normal Distribution")
plt.xlabel("Data Points")
plt.ylabel("Probability")
plt.show()

#### Keeping Standard deviation constant = 3 and varying mean
plt.figure(figsize=(10,7))

norm_mean = [0,2,5]
standard_deviation = 3
for mean in norm_mean:
  plt.plot(x,sp.norm.pdf(x,mean,sd),label="Mean = {} & SD = {}".format(mean,sd))

plt.legend()
plt.title("Probability Density Function(PDF) of Normal Distribution")
plt.xlabel("Data Points")
plt.ylabel("Probability")
plt.show()
```

##**Q1--- Part:4) Uniform (continuous) Distribution**

```
n = 1000
start = [0,1.5,0.5]
width = [1,1.5,2]
for i in range(0,len(start)):
  x = sp.uniform.rvs(size=n, loc = start[i], scale=width[i])
  plt.plot(x,sp.uniform.pdf(x,loc = start[i], scale=width[i]),label="a = {} & b = {}".format(start[i],start[i]+width[i]))
  print("Mean = {}, Variance = {}\n".format(sp.uniform.mean(loc=start[i],scale=width[i]),sp.uniform.var(loc=start[i],scale=width[i])))
plt.legend()
plt.title("Probability Density Function(PDF) of Uniform Distribution")
plt.xlabel("Data Points")
plt.ylabel("Probability")
plt.ylim(0,1.2)
plt.show()


### Mean = (a+b)/2
### Variance = (1/12)*(b-a)^2
```

```
n = 1000
start = [0,1.5,0.5]
width = [1,1.5,2]
for i in range(0,len(start)):
  y = np.arange(-1,4,0.01)
  plt.plot(y,sp.uniform.cdf(y,loc = start[i], scale=width[i]),label="a = {} & b = {}".format(start[i],start[i]+width[i]))
plt.legend()
plt.title("Cumulative Distribution Function (CDF) of Uniform Distribution")
plt.xlabel("Data Points")
plt.ylabel("Probability")
plt.ylim(0,1.2)
plt.show()
```
